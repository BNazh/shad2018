```python
import matplotlib.pyplot as plt
import skimage
from skimage import  io
import numpy as np
```

# image

```python
def show(image, title=None):
    plt.imshow(image)
    if title is not None:
        plt.title(title)
    plt.show()
```

```python
im = io.imread('./duckling.png')
```

```python
show(im, 'Duckling')
```

```python
im.shape
```

```python
up = im[:im.shape[0] // 2]
```

```python
down = im[im.shape[0] // 2:]
```

```python
show(up, 'up')
```

```python
show(down, 'down')
```

```python
plt.imsave('up.png', up)
```

```python
plt.imsave('down.png', down)
```

```python
im.shape
```

```python
for ch in range(3):
    l = [0, 1, 2, 3]
    l.remove(ch)
    show(im[:, :, ch], ['red', 'green', 'blue'][ch])
    show(im[:, :, l], 'all - %s' % ['red', 'green', 'blue'][ch])
```

```python
mean = im.mean(axis=-1)
```

```python
gray = skimage.color.rgb2gray(im) * 256
```

```python
show(np.concatenate((mean, gray.astype(float)), axis=1))
```

```python
hsv = skimage.color.rgb2hsv(im[:, :, :3])
```

```python
show(np.concatenate([hsv[:, :, i] for i in range(3)], axis=1))
```

```python
show(1. - gray)
```

```python
show(im[:, ::-1, :])
```
